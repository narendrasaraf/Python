## Assignment no: 3.Write a Python program to encrypt a message using RSA. Explain the concept of public and private keys.

```python
from Crypto.PublicKey import RSA
from Crypto.Cipher import PKCS1_OAEP

key = RSA.generate(2048)  # 2048-bit key
public_key = key.publickey()
private_key = key

message = b"Hello, RSA Encryption!"

cipher = PKCS1_OAEP.new(public_key)
encrypted_message = cipher.encrypt(message)
print(f" Encrypted Message: {encrypted_message}")

decrypt_cipher = PKCS1_OAEP.new(private_key)
decrypted_message = decrypt_cipher.decrypt(encrypted_message)
print(f" Decrypted Message: {decrypted_message.decode()}")
```
