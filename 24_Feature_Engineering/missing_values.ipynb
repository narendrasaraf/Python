````python
# # Handling Missing Values

# To replace missing values in a dataset, you can use the `fillna()` method in pandas. Hereâ€™s an example:

# ```python
import pandas as pd

# Example DataFrame with missing values
df = pd.DataFrame({
    'A': [1, 2, None, 4],
    'B': [5, None, None, 8]
})

# Replace missing values with a specific value (e.g., 0)
df_filled = df.fillna(0)

print(df_filled)
````
